```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 1000

d = {
    "tfidfScore":       np.round(np.random.uniform(0, 30, n), 3),   # matches real range
    "skillScore":       np.round(np.random.uniform(0, 100, n), 3),
    "matchedCount":     np.random.randint(0, 34, n),
    "missingCount":     np.random.randint(0, 34, n),
    "experienceYears":  np.round(np.random.uniform(0, 12, n), 1),
    "programmingScore": np.round(np.random.uniform(0, 100, n), 2),
    "backendScore":     np.round(np.random.uniform(0, 100, n), 2),
    "frontendScore":    np.round(np.random.uniform(0, 100, n), 2),
    "databaseScore":    np.round(np.random.uniform(0, 100, n), 2),
    "mlAiScore":        np.round(np.random.uniform(0, 100, n), 2),
    "dataScienceScore": np.round(np.random.uniform(0, 100, n), 2),
    "cloudScore":       np.round(np.random.uniform(0, 100, n), 2),
    "devopsScore":      np.round(np.random.uniform(0, 100, n), 2),
    "toolsScore":       np.round(np.random.uniform(0, 100, n), 2),
}

df = pd.DataFrame(d)

# Label: simulate finalScore = skillScore*0.6 + tfidfScore*0.4 + expBoost
exp_boost = np.minimum(df["experienceYears"], 5) * 2
final_score = df["skillScore"] * 0.6 + df["tfidfScore"] * 0.4 + exp_boost

df["label"] = (final_score > 30).astype(int)  # shortlist threshold = 30%

print(df["label"].value_counts())
df.to_csv("ats_training_dataset_1000.csv", index=False)
df.head()
```
